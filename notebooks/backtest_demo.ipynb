# Portfolio Backtester Demo

This notebook loads deterministic local fixtures and runs the Phase 1 API.

```python
import pandas as pd
from portfolio_bt.api import compare_portfolios, run_backtest

prices = pd.read_parquet("tests/fixtures/prices/compare_2011_2020.parquet")
matrix = prices.pivot_table(index="date", columns="ticker", values="adj_close", aggfunc="last").sort_index()
```

```python
single = run_backtest(
    matrix,
    {"VTI": 0.60, "VXUS": 0.20, "BND": 0.20},
    start="2011-02-01",
    end="2020-12-31",
    rebalance="annual",
    portfolio_name="Three Fund",
)
single["metrics"]
```

```python
comparison = compare_portfolios(
    matrix,
    [
        {"name": "Three Fund", "weights": {"VTI": 0.60, "VXUS": 0.20, "BND": 0.20}},
        {"name": "US Only", "weights": {"VTI": 1.0}},
    ],
    benchmark={"SPY": 1.0},
)
comparison["comparison_table"]
```
